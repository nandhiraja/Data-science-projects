```python
# import the nessceary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder,MinMaxScaler
from sklearn.tree import DecisionTreeRegressor
import xgboost as xgb
```

```python
# Split the x and y value

x=data[["Avg_Price","Offline_Spend","Online_Spend","Quantity"]]                   
y= data["GST"]
```

```python
# split the x and y train and test data 

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=43)
```

```python
# scaling the data

feature  =  StandardScaler()

feature_data = feature.fit_transform(x_train)  # fit and transform  the data

feature_test =  feature.transform(x_test)
```

```python
# Initialize the models 

model_tree = DecisionTreeRegressor(criterion="poisson",random_state=6,max_depth=10,ccp_alpha=0.0)

model_xgb = xgb.XGBRegressor( learning_rate = 0.9,
                max_depth = 75, alpha = 1, n_estimators = 10)
```

```python
# fit the train data into the model

model_tree.fit(feature_data,y_train)
```

```python
# fit the train data into the model

model_xgb.fit(feature_data,y_train)

```

```python
# make the prediction in the train dataset

tree_train_score = model_tree.predict(feature_data)

XGB_train_score = model_xgb.predict(feature_data)
```

```python
# make the prediction based in test data

tree_test_score = model_tree.predict(feature_test)

XGB_test_score = model_xgb.predict(feature_test)
```

```python
# import the accuracy of the model
from sklearn import metrics

#  -------------------  R2 score ----------------

score_tree= metrics.r2_score(y_train, tree_train_score)              # train scores
score_XGB= metrics.r2_score(y_train, XGB_train_score)

score_tree_test = metrics.r2_score(y_test, tree_test_score)          # test scores
score_XGB_test = metrics.r2_score(y_test, XGB_test_score)
```

```python
#  ----------------- Mean absolute error ------------

mean_score_tree  =  metrics.mean_absolute_error(y_train,tree_train_score)         # train data
mean_score_XGB  =  metrics.mean_absolute_error(y_train,XGB_train_score)


mean_score_tree_test = metrics.mean_absolute_error(y_test,tree_test_score)        # test data
mean_score_XGB_test = metrics.mean_absolute_error(y_test,XGB_test_score)
```

```python
print(f"---- R2score ----\nD.Tree   {score_tree}\nXGB      {score_XGB}\n")

print(f"-----R2score ----\nD.tree  {score_tree_test}\nXGB      {score_XGB_test}")
```

```python
print(f" Mean Absolute Error \n\nD.TREE :   {mean_score_tree}\nXGB  :     {mean_score_XGB}\n")

print(f" Mean Absolute Error \n\nD.TREE :   {mean_score_tree_test}\nXGB  :    {mean_score_XGB_test}")
```
